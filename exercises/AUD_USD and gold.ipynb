```python
result = pd.concat([gold, AUD_USD], axis=1).reindex(gold.index)
result.head()
```

```python
result1 = result['Adj Close']
result1.head()
```

```python
result1.columns = ['% change gold', '% change currency']
```

```python
# find where gold is negative
# print the next row of change in currency

#column_var -- column which changes
# column1 -- column which is bought/sold

def backtest(df,column_var, column1):
    a = (df[df[column_var]>0][column1].shift(periods = -1, freq = None,  axis = 0,))
    print( pd.DataFrame(a))
    

    print(a.head())
    print( '%return on investment:',a.sum(axis=0)    )  

backtest(comparison , '% change gold', '% change currency')
```

```python
#graph to show minimum %change to invest
#short one column, buy the other

def graph(df,column_var, column1):
    array = np.arange(-0.1,0.1,0.01)
    for i in array:
        b = pd.DataFrame( (df[df[column_var]>i][column1].shift(periods = 1, freq = None,  axis = 0,))  )
    
        i += i 
        print( i, b.sum(axis = 0))
        

graph(comparison , '% change gold', '% change currency')
    
```
